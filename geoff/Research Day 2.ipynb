```python
import afterglowpy as grb
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Array of times in seconds
t = np.array([1.0e3, 1.0e4, 1.0e5, 1.0e6])

# Array of frequencies in Hz
nu = np.array([1.0e17, 1.0e17, 1.0e17, 1.0e17])

Z = {'jetType': grb.jet.TopHat,  # type of jet structure
     'specType': 0,              # type of spectrum
     'thetaObs': 0.0,   # observer angle (radians)
     'E0': 1.0e53,      # isotropic energy (erg)
     'thetaCore': 0.1,  # jet opening angle (radians)
     'n0': 1.0,         # ambient ISM density (cm^{-3})
     'p': 2.2,          # electron energy slope (no units)
     'epsilon_e': 0.1,  # fraction of energy in accelerated electrons (no units)
     'epsilon_B': 0.01, # fraction of energy in magnetic field (no units)
     'ksiN': 1.0,       # fraction of electrons that get accelerated (no units)
     'dL': 1.0e28,      # luminosity distance (cm)
     'z': 0.5           # redshift (no units)
    } 

Fnu = grb.fluxDensity(t, nu, **Z)
```

```python
plt.plot(t, Fnu, marker='.')
plt.xscale('log')
plt.yscale('log')
```

```python
# Make a light curve
t = np.geomspace(1.0e3, 1.0e6, 100)
nu = np.empty(t.shape)
nu[:] = 1.0e17

Fnu = grb.fluxDensity(t, nu, **Z)

plt.plot(t, Fnu)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('time (s)')
plt.ylabel('Flux (mJy)')
```

```python
# Make a spectrum
nu = np.geomspace(1.0e6, 1.0e20, 100)   #Hz

t = np.empty(nu.shape)
t[:] = 1.0e4  # seconds

Fnu = grb.fluxDensity(t, nu, **Z)

plt.plot(nu, Fnu)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('frequency (Hz)')
plt.ylabel('Flux (mJy)')
```
